This project analyzes bank loan data to evaluate lending performance, portfolio health, and borrower risk characteristics. The goal is to build an end-to-end analytics workflow using Excel for data cleaning, SQLite and Python for exploratory data analysis (EDA), and Power BI for interactive dashboard reporting.

The analysis focuses on key business metrics including loan applications, funded amounts, repayment performance, borrower characteristics, and loan risk segmentation. The project also distinguishes between good and bad loans to assess overall portfolio quality and identify risk patterns.

This project simulates a real-world banking analytics use case and demonstrates practical skills in SQL, Python, financial analytics, and business intelligence reporting.

## Tools Used
* **Excel** — Data Cleaning
* **SQLite** — Data Storage & SQL Queries
* **Python** (Pandas, Matplotlib, Seaborn) — Data Analysis
* **Microsoft Power BI** — Dashboard Reporting

##Project Objectives
1. Analyze loan application trends
2. Evaluate loan repayment performance
3. Identify high-risk borrower segments
4. Measure portfolio health using KPIs

```python
import pandas as pd
import sqlite3

db_path = '/content/drive/MyDrive/Bank_Loan_Project/bank_loan.db'

# 3. Load data
url = 'https://raw.githubusercontent.com/mutumaian29/Bank-Lending-Intelligence-Dashboard/refs/heads/main/financial_loan_data_excel.xlsx%20-%20financial_loan.csv'
df = pd.read_csv(url)

# 4. Connect and Save
conn = sqlite3.connect(db_path)
df.to_sql("loans", conn, if_exists="replace", index=False)
conn.close()

print(f"Database saved successfully to: {db_path}")
```
